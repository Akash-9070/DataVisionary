```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x= np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

# functional method of ploting

```python
plt.plot(x,y)
plt.xlabel("X LABEL")
plt.ylabel("Y label")
plt.title("Title")
plt.show
```

```python
# fig 1
plt.subplot(1,3,1)
plt.plot(x,y,"r") 
# fig 2
plt.subplot(1,3,2)
plt.plot(x,y,"b")
# fig 3
plt.subplot(1,3,3)
plt.plot(x,y*2,"g")
```

# ploting through the object oriented method 


```python
fig  = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.4])

axes1.plot(x,y)
axes1.set_title("larger plot")

axes2.plot(y,x)
axes2.set_title("smaller plot ")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
# PLOTING SUB PLOTS BY OOPs
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

# # we acn also iterate through it 
# for current_ex in axes:
#     current_ex.plot(x,y)
    
# or we can also get elements through it 
axes[0].plot(x,y)
axes[0].set_title("A")

axes[1].plot(y,x)
axes[1].set_title("B")

plt.tight_layout()
```

# Figure size and DPI


```python
fig = plt.figure(figsize = (3,2)) # IT SAYS THAT FIG WILL BE 3/2 INCHES IN SIZE

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize=(6,2)) # in inches.

axes[0].plot(x,y)
                           
axes[1].plot(y,x)

plt.tight_layout() #  to avoid overlap
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2 ,label = "X square")
ax.plot(x,x**3 ,label = "X cubed")

ax.legend(loc=0) # loc =0 for it choose the best location for it 
```

# Customizing the plots

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="red",lw = 3 , ls = "-.")
# we can also add the "RGB Hex codes" to custom the color
```

# markers

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="red",lw = 1 , ls = "-.", marker = "o", markersize = 20
       ,markerfacecolor="yellow",markeredgewidth=3,markeredgecolor="green")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="red",lw = 3 , ls = "-.")


# to decide the range of the plot i want to see..
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
