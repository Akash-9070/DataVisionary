```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()

```

```python
flights.head()
```

# Matrix plot

heat maps

```python
columns_to_drop = [ "sex","smoker","day","time"]
num_data = tips.drop(columns=columns_to_drop)
tc = num_data.corr()

sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
fp = flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(fp,cmap="Blues")

# some of the colormap options are
# cividis
# plasma
# viridis
# magma
```

```python
sns.clustermap(fp)
```

# Grids


```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset("iris")
iris.head()
```

```python
# PairGrid gives more control over the plots then the pairplot

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col="time",row="smoker")
g.map(sns.distplot,"total_bill")
```
